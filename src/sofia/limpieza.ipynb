# Limpieza


```python
import pandas as pd
import unicodedata
import re

### ───────────── FUNCIONES DE LIMPIEZA ─────────────

def quitar_acentos(texto):
    if isinstance(texto, str):
        return unicodedata.normalize('NFKD', texto).encode('ascii', errors='ignore').decode('utf-8')
    return texto

def convertir_minusculas(texto):
    if isinstance(texto, str):
        return texto.lower()
    return texto

def eliminar_caracteres_especiales(texto):
    if isinstance(texto, str):
        texto = texto.replace('"', '').replace("'", '')
        texto = re.sub(r'[\r\n\t]', ' ', texto)
        return texto
    return texto

def limpiar_espacios(texto):
    if isinstance(texto, str):
        texto = re.sub(r'\s+', ' ', texto)
        return texto.strip()
    return texto

def reemplazar_comas_por_pipes(texto):
    """Reemplaza todas las comas por pipes."""
    if isinstance(texto, str):
        return texto.replace(',', '|')
    return texto

### ───────────── FUNCIONES MAESTRAS POR COLUMNA ─────────────

def limpiar_campo_general(valor):
    """Limpia un campo común (sí reemplaza comas por pipes)."""
    return limpiar_espacios(
        reemplazar_comas_por_pipes(
            eliminar_caracteres_especiales(
                convertir_minusculas(
                    quitar_acentos(valor)
                )
            )
        )
    )

def limpiar_direccion(valor):
    """Limpia la dirección pero NO reemplaza comas."""
    return limpiar_espacios(
        eliminar_caracteres_especiales(
            convertir_minusculas(
                quitar_acentos(valor)
            )
        )
    )

### ───────────── PROCESAMIENTO COMPLETO ─────────────

archivo_entrada = "muestra_basicos_todos.csv"
archivo_salida = "muestra_basicos_limpio.csv"

# Asignamos columnas si no tiene header
columnas = [
    "producto", "presentacion", "marca", "tipo", "catalogo",
    "precio", "fecha", "tienda", "tipo_tienda", "sucursal",
    "direccion", "estado", "ciudad", "latitud", "longitud"
]

# 1. Leer todo el archivo
df = pd.read_csv(archivo_entrada, header=None, names=columnas)

# 2. Aplicar limpieza columna por columna
for col in df.columns:
    if col == "direccion":
        df[col] = df[col].apply(limpiar_direccion)
    else:
        df[col] = df[col].apply(limpiar_campo_general)

# 3. Eliminar columnas latitud y longitud
df = df.drop(['latitud', 'longitud'], axis=1)

# 4. Guardar el archivo limpio
df.to_csv(archivo_salida, index=False, sep='|')

print(f"✅ Limpieza total hecha, 'direccion' conserva comas y columnas 'latitud' y 'longitud' fueron eliminadas.")
```
